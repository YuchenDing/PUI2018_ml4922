## 1. Chi Sq Distribution

```python
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
distributions = ['chisq']
```

```python
mymean = 100
```

```python
# the chi sq distribution takes the number of degrees of freedom, which luckily is the same as the mean!
df = mymean
```

```python
# a single chi sq distribution
import numpy as np
md = {}
# md is an empty disctionary
md['chisq'] = np.random.chisquare(df, size=100)
# md now is a dictionary with one object: an array by the key 'chisq'
pl.hist(md['chisq'], bins = 30)
pl.title('Histogram of a single chi sq distribution')
pl.ylabel('N')
pl.xlabel('x')
```

### Figure 1: Chi sq distribution: df = 100, N=100

```python
# let numpy calculate what the mean and standard deviation of this distribution are
print ("Chisq mean: %.2f, standard deviation: %.2f"%(md['chisq'].mean(), md['chisq'].std()))
```

```python
# set distribution of different size(all integers between10 and 2000)
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
# mysize = 1000 / (np.array(range(1, 100) + [10]))
# mysize = (np.random.rand(100) * 1000).astype(int)
print (mysize, mysize.shape)
```

```python
# I am gonna overwrite the disctionary entry created into a nested dictionary 
# the outside dictionary hosts the distributions as dictionary, 
# and each distribution dictionary contains the 100 distribution (arrays of diferent sizes)
md['chisq'] = {} 

#and do it in a for loop. not pythonic, but easily readable

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
# save the means for plotting later    

md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

### Figure 2: The plot of Chi Sq distribution sample mean against sample size. As sample size increases, the sample mean is close to 100, the population mean we set.

```python
from scipy.stats import norm
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(12, 10))
n,bins,patches=plt.hist(allmeans,30,normed=1,facecolor='steelblue',align='mid')
(mu, sigma) = norm.fit(allmeans)
y = mlab.normpdf(bins, mu, sigma)
pl.plot(bins, y, 'r-', linewidth=2)

pl.grid(True)
pl.title('Histogram of the sample means of Chi Sq',fontsize=20)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('frequency', fontsize = 18)
```

### Figure 3: The histogram of the sample means of Chi Sq. This shows the distribution of all sample means is close to guassian.
